```python
import pandas as pd
import numpy as np

def clean_dataset(df):
    assert isinstance(df, pd.DataFrame), "df needs to be a pd.DataFrame"
    df.dropna(inplace=True)
    indices_to_keep = ~df.isin([np.nan, np.inf, -np.inf]).any(axis=1)
    return df[indices_to_keep].astype(np.float64)
    
```

```python
df['price'] = df['price'].replace((np.inf, -np.inf, np.nan), 0).reset_index(drop=True)
```

```python
df['normalized-losses'] = df['normalized-losses'].replace((np.inf, -np.inf, np.nan, "?"), df['normalized-losses'].mean()).reset_index(drop=True)
```

```python
df['bore'] = df['bore'].replace((np.inf, -np.inf, np.nan, "?"), df['bore'].mean()).reset_index(drop=True)
```

```python
X = df_num.drop('price', axis=1)
y= df_num['price']
```

```python
seed= 123
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=seed)
```

```python
print(X_train.shape, X_test.shape, y_train.shape,  y_test.shape)
```

```python
X_train = X_train.replace((np.inf, -np.inf, np.nan), 0).reset_index(drop=True)
```

```python
y_train = y_train.replace((np.inf, -np.inf, np.nan), 0).reset_index(drop=True)
```

Model de Regression lineaire multiple

```python
lineReg = LinearRegression()
lineReg.fit(X_train, y_train)
```

```python
def evaluted_model(model):
    trains_pred = model.predict(X_train)
    test_pred = model.predict(X_test)
    rmse_train= mean_squared_error(y_train, trains_pred, squared=False)
    rmse_test= mean_squared_error(y_test, test_pred, squared=False)
    return rmse_train, rmse_test
```

```python
#Evaluation du model regression lineaire
evaluted_model(lineReg)
```

Random Forest

```python
random_forest= RandomForestRegressor(random_state= seed)
random_forest.fit(X_train, y_train)
```

```python
#Evaluation du model Random Forest
evaluted_model(random_forest)
```

On constate que le model de RandomForest est une meilleur performance du point de vue erreur quadratique. 
Donc le model Random forest reste meilleur pour la suite de l'implementation
